# **Visualizacion de Datos**

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(tiendas, ventas)
plt.title("Facturación total por tienda")
plt.xlabel("Tiendas")
plt.ylabel("Ventas totales")
plt.ylim(1000000000.0, 1200000000.0)  # rango del eje Y
plt.show()
```

```python
plt.bar(tiendas, calificacion)
plt.title("calificacion por tienda")
plt.xlabel("Tiendas")
plt.ylabel("calificacion promedio")
plt.ylim(3.9, 4.1)   # ajusta el rango
plt.show()
```

```python
plt.bar(tiendas, envio)
plt.title("Costo de Envio por tienda")
plt.xlabel("Tiendas")
plt.ylabel("Costo Envio")
plt.ylim(20000, 30000.0)   # ajusta el rango
plt.show()
```
